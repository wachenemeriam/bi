```python
#plot 
plt.plot(df['Article IT,'],df['Total Quantity,'], c='violet' )# quantiter total de chaque article
plt.xlabel('Article IT,')
plt.ylabel('Total Quantity,')
plt.title('total quantiter of each item')
plt.show()
```
